# Generate simulation data

This notebooks allows to generate synthetic data to test the segmentation and post-processing. In particular it allows to create spots distributions with various levels of asymetry.

```python
from napari_cellspots.simulation import simulate_image
import napari
import pyics
import numpy as np
import tifffile
import os
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
#location
data_location = '../sample_data/simulations'
# Number of images
n_images = 2
# Levels of asymetry
factors = [0,1]
# naming of files
names = ['sym', 'asym']
```

```python
data_location = Path(data_location)

for n in range(len(names)):

    cur_path = data_location / f'simulation_{names[n]}'
    os.makedirs(cur_path, exist_ok=True)

    for i in range(n_images):

        simul = simulate_image(symmetry_factor=factors[n])
        simul = np.flip(simul, axis=0)
        simul = (255*(simul / simul.max())).astype(np.uint8)
        full_simul = np.zeros((simul.shape[0], 7, simul.shape[1], simul.shape[2]), dtype=np.uint8)
        full_simul[:, full_simul.shape[1] // 2] = simul.copy()
        tifffile.imwrite(Path(cur_path).joinpath(f'simul_{names[n]}_{i}.tiff'), full_simul)
```
